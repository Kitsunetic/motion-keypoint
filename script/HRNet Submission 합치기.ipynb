```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
dpath = Path("results/hrnet-train2/example/HRNet-W48-ce-512x512-plus_augment-maw-rr2.0-ReduceLROnPlateau")
```

```python
csv_files = [
    "HRNet-W48-ce-512x512-plus_augment-maw-rr2.0-ReduceLROnPlateau-256x256_512x512_1.csv",
    "HRNet-W48-ce-512x512-plus_augment-maw-rr2.0-ReduceLROnPlateau-256x256_512x512_2.csv",
    "HRNet-W48-ce-512x512-plus_augment-maw-rr2.0-ReduceLROnPlateau-256x256_512x512_3.csv",
    "HRNet-W48-ce-512x512-plus_augment-maw-rr2.0-ReduceLROnPlateau-256x256_512x512_4.csv",
    "HRNet-W48-ce-512x512-plus_augment-maw-rr2.0-ReduceLROnPlateau-256x256_512x512_5.csv",
]
```

```python
csvs = [pd.read_csv(dpath / f) for f in csv_files]
```

```python
csv = csvs[0]
```

```python
csv
```

```python
datas = [csv.iloc[:, 1:].values.astype(np.float) for csv in csvs]
```

```python
datas[0].shape
```

```python
data_stack = np.stack(datas)
```

```python
data_stack.shape
```

```python
data_median = np.median(data_stack, 0)
```

```python
data_median
```

```python
data_mean = np.mean(data_stack, 0)
```

```python
files_np = csvs[0].image.values
```

```python
files_np
```

```python
df_median = np.concatenate([np.expand_dims(files_np, 1), data_median], 1)
```

```python
df_mean = np.concatenate([np.expand_dims(files_np, 1), data_mean], 1)
```

```python
df_median
```

```python
df_mean = pd.DataFrame(df_mean, columns=csvs[0].columns)
df_mean
```

```python
df_median = pd.DataFrame(df_median, columns=csvs[0].columns)
```

```python
df_median
```

```python
df_mean.to_csv(
    dpath / "HRNet-W48-ce-512x512-plus_augment-maw-rr2.0-ReduceLROnPlateau-256x256_512x512_12345mean.csv",
    index=False,
)
```

```python
df_median.to_csv(
    dpath / "HRNet-W48-ce-512x512-plus_augment-maw-rr2.0-ReduceLROnPlateau-256x256_512x512_12345median.csv",
    index=False,
)
```
